```python
# Use this cell to begin your analyses, and add as many cells as you would like!
# Import packages needed 
import pandas as pd 
import numpy as np
```

```python
lego_sets = pd.read_csv('lego_sets.csv')
lego_sets
```

```python
lego_sets.count()
# We have 11986 observations but 11833 set_num. 
#We have to drop any missing observation and check for duplicate set_num

lego_sets1 = lego_sets.dropna(subset= ['set_num'])
```

```python
#Check for duplicate rows
dupboolean = lego_sets1.duplicated()
duplicaterows=lego_sets1[lego_sets1.duplicated()]

print(dupboolean)
print(duplicaterows)
```

```python
#read in parent themes dataset
parent_themes = pd.read_csv('parent_themes.csv')
parent_themes.count()
```

```python
#merge both datasets together
lego_full = lego_sets1.merge(parent_themes, left_on='parent_theme', right_on='name',
                            how='outer')
lego_full1=lego_full.rename(columns= {'name_x': 'name'}).drop('name_y', axis=1)
lego_full1.count()
```

```python
#we are going to create a dataframe of only starwars dataset
star_wars=lego_full1.loc[(lego_full1['parent_theme'] == 'Star Wars') & 
                         (lego_full1['is_licensed'] == True)]
star_wars.count()
```

```python
# Filter the lego_full1 dataframe to include only rows where 'is_licensed' is True
lego_full2 = lego_full1.loc[lego_full1['is_licensed'] == True]

# Count the number of non-NA/null observations in each column of the filtered dataframe
lego_full2.count()
```

```python
print(star_wars['parent_theme'].count())
print(lego_full2['parent_theme'].count())
jedi = star_wars['parent_theme'].count()
non_jedi = lego_full2['parent_theme'].count()
the_force = (jedi/non_jedi) * 100
print(the_force)
```

```python
lego_full3 = lego_full2.groupby(['year', 'parent_theme']).count().reset_index()
lego_full3
```

```python
lego_max = lego_full3.groupby(['year'])['set_num'].max().reset_index()
lego_max1=lego_max.rename(columns={'set_num': 'max'})
lego_max1
```

```python
lego_full4=lego_full3.merge(lego_max1, on='year')
lego_full4
```

```python
lego_full5=lego_full4.loc[(lego_full4['parent_theme'] =='Star Wars') & 
               (lego_full4['set_num'] < lego_full4['max']), ['year']]
lego_full5
```

```python
new_era=lego_full5.loc[81,['year']]
new_era=int(new_era)
new_era
```
